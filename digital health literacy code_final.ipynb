```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create the figure and subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 10))  # 1 row, 3 columns

# Plot the first graph
sns.countplot(x='HaveDevice_SmartPh', data=df, palette='viridis', ax=axs[0])
axs[0].set_title('Number of people using Smartphones')
axs[0].set_xlabel('Smartphone Ownership Status')
axs[0].set_ylabel('Count')
axs[0].set_xticklabels(axs[0].get_xticklabels(), rotation=45, ha='right', fontsize=10)

# Plot the second graph
sns.countplot(x='UseInternet', data=df, palette='magma', ax=axs[1])
axs[1].set_title('Number of People having good internet connection')
axs[1].set_xlabel('Internet Usage Status')
axs[1].set_ylabel('Count')
axs[1].set_xticklabels(axs[1].get_xticklabels(), rotation=45, ha='right', fontsize=10)

# Plot the third graph
sns.countplot(x='WearableDevTrackHealth', data=df, palette='deep', ax=axs[2])
axs[2].set_title('Number of People tracking health with wearables')
axs[2].set_xlabel('Another Variable Status')
axs[2].set_ylabel('Count')
axs[2].set_xticklabels(axs[2].get_xticklabels(), rotation=45, ha='right', fontsize=10)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
ax = sns.countplot(x='Age', hue='UseInternet', data=df, palette='Set2')

# Rotate x-labels
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=12)

# Remove legend
ax.legend_.remove()

plt.xlabel('Age', fontsize=14)
plt.ylabel('UseInternet', fontsize=14)  # Adding y-axis label
plt.title('Age Vs Use of internet', fontsize=16)

# Add legend to the right side
plt.legend(title='X axis labels', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
